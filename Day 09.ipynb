# 결측치

```python
# 결측치 : 비어있는 값. 분석 결과 왜곡시킴 / 처리 = 1. 행 처리 2. 값을 채워넣기(평균값 or 예측값)


```

```python
age = pd.Series([26, 42, 27, 25, 20, None])

score = pd.Series([3.8, 4.2, 2.5, 3.0, 4.0, 2.1])

salary = pd.Series([2600, 4200, 2700, 2500, 2000, 2100])

stu_class = pd.Categorical([1,2,2,3,4,None])

gender = pd.Categorical(['F','F','M','M',None,None])

df = pd.DataFrame( {'age' : age,
                   'score' : score,
                   'salary' : salary,
                   'stu_class' : stu_class,
                   'gender' : gender} )

df
```

```python
df_new = df.copy()           # df복제

pd.isna(df_new)              # 데이터프레임의 모든 값이 boolean 형태로 표시되도록 하며, nan인 값에만 True가 표시되게 하는 함수

df_new.isnull().sum()        # 열별로 결측치 갯수 확인 
```
