```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# astype 을 통해 형변환 가능
#titanic['Survived'] = titanic['Survived'].astype(object)

titanic = pd.read_csv("ds7/data/titanic.csv")
```

```python
titanic.info()
titanic.head()
```

```python
titanic.isnull().sum()
titanic.drop(columns=['Cabin'], inplace=True)
titanic['Survived'] = titanic['Survived'].astype(object)
titanic['Pclass'] = titanic['Pclass'].astype(object)
```

```python
titanic.dropna(inplace=True)
titanic.isnull().sum()
```

```python
plt.bar((0,1),titanic['Survived'].value_counts())
plt.xticks((0,1),('Dead','Alive'))
```

```python
category_feature = [col for col in titanic.columns if titanic[col].dtypes=='object']

category_feature = list(set(category_feature) - set(['Survived','PassengerId']))
```

```python
for col in category_feature :
    titanic[col].value_counts().plot(kind='bar')
    
    plt.title(col)
    plt.show()
```

```python
def bar_chart(feature) :
    alive = titanic[titanic['Survived']==1][feature].value_counts()
    dead = titanic[titanic['Survived']==0][feature].value_counts()
    df = pd.DataFrame([alive,dead])
    df.index = ['Alive', 'Dead']
    df.plot(kind= 'bar' , stacked = False , figsize=(10,5))              # stacked = 누적
```

```python
bar_chart('Pclass')
```

```python
bar_chart('Sex')
```

```python
bar_chart('Embarked')
```

```python
bar_chart('Parch')
```

```python
bar_chart('SibSp')
```
