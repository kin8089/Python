# 2006년 ~ 2022년 / 입력한 월간 분실물 총합 비교

```python
i = int(input('비교 할 개수를 선택하세요'))

num_list = (range(i))
monthlist = []
namelist = []

print(num_list)

for i in num_list:
    a = int(input('분실물을 보고싶은 년도를 기재하세요.'))
    b = int(input('분실물을 보고싶은 월을 기재하세요.'))
    if b < 10 :
        b = '0' + str(b) 
    
    start = str(a)+'-'+str(b)+'-01'
    end = str(a)+'-'+str(b)+'-31'
    
    df_y = df[df['등록일자'].between(start,end)]
    df_z = df_y['분실물종류'].value_counts().rename_axis('분실물').reset_index(name = '합계')
    df_q = df_z['합계'].sum()
    monthlist.append(df_q)
    v = (str(a)+'년'+str(b)+'월')
    namelist.append(v)


plt.figure(figsize=(10,10))
plt.ylabel('월별 분실물 총합')

plt.plot(namelist, monthlist,'o-')
plt.grid()
```
