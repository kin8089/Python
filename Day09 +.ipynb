# DataFrame 합치기

```python
fruit = pd.DataFrame(
        {'Num' : [123, 456, 789, 1011, 1112],
         'Fruit' : ['Apple', 'Banana', 'Strawberry', 'Peach', 'Melon']})
         
grade = pd.DataFrame(
        {'Num' : [123, 789, 1314],
         'Grade' : ['A', 'B', 'C']})

price = pd.DataFrame(
        {'No' : [123, 789, 1314],
         'Grade' : ['100', '200', '500']})
```

```python
fruit

grade

price
```

```python
df = pd.concat([fruit,grade,price])         # concat 은 그냥 데이터 양을 늘리는 방식.

df
```

```python
df = pd.merge(fruit,grade, on = 'Num', how = 'left')   # Inner join : how = 'left'
df

df.isnull().sum()
```

```python
df = pd.merge(fruit,price,left_on='Num',right_on='No',how='left')

df
```

```python
df = pd.read_csv('data/dirtydata.csv')
```

```python
x = df['Calories'].mean()

df['Calories'] = df['Calories'].fillna(x)  

df['Date'] = df['Date'].fillna(value='2020/12/22') 

#df = df[df['Duration'].isin([30, 45, 60])]        # 해당 값이 아닌것 다 제거

df.loc[7, 'Duration'] = 45                         # 특정 값 수정

df['Date'] = pd.to_datetime(df['Date'])            # 날짜 형식을 통째로 교정

df.duplicated()                                    # 중복여부 확인

df.drop_duplicates(inplace=True)                   # 중복 삭제

df
```

```python
pd.get_dummies(df['Duration'])          # 데이터 전처리 / 가변수 만들기
```

```python
df['MaxP'] = pd.cut(x = df['Maxpulse'], bins=[100, 140, 180], labels=['0','1'])
df['Cal'] = pd.cut(x = df['Calories'], bins=[190, 300, 500], labels=['0','1'])
df
```
