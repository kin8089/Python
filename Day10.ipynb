```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 10 , 1])
ypoints = np.array([1, 10 , 1])

plt.plot(xpoints,ypoints, 'r.:')
plt.plot(xpoints,ypoints*2, color='springgreen', marker='*', linewidth=2)
plt.xlabel('Kinds')
plt.ylabel('Count')
plt.title('king')
plt.grid()
plt.xticks(np.arange(1, 10))
plt.yticks(np.arange(1, 21))
plt.legend(['A','B'])
plt.show()
```

```python
df = pd.read_csv('ds7/data/pop_Seoul.csv')

df
```

```python
fig = plt.figure(figsize=(10, 10))

plt.plot(df.loc[:, 'year'], df.loc[:,'population'], 'go:')

plt.title("Seoul Population Change")

plt.xticks(df.loc[:, 'year'], labels=df.loc[:, 'year'])
plt.ylabel("10Million")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))

plt.bar(df.loc[:, 'year'], df.loc[:,'population'], width=3)

plt.title("Seoul Population Change")

plt.xticks(df.loc[:, 'year'], labels=df.loc[:, 'year'])
plt.ylabel("10Million")
plt.show()
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

df = pd.read_csv('ds7/data/pop_Seoul2.csv', names=['year','seoul','capital'],skiprows=[0])

fig = plt.figure(figsize=(10, 10))

year = df.loc[:,'year']
seoul = df.loc[:,'seoul']
capital = df.loc[:,'capital']

plt.plot(year,seoul, 'go:', label='서울')

plt.plot(year,capital, 'ro:', label='경기도')

plt.xticks(range(12), year, rotation=45)

plt.title("인구 변화")
plt.ylabel("퍼센트")

plt.legend()
```
